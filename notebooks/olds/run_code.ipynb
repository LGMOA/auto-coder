```python
from PIL import Image, ImageDraw

def round_corners(image_path, output_path, corner_radius=90):
    # 打开原始图片
    img = Image.open(image_path).convert("RGBA")
    
    # 创建一个新的带有透明背景的图片
    width, height = img.size
    mask = Image.new('L', (width, height), 0)
    
    # 绘制圆角
    draw = ImageDraw.Draw(mask)
    draw.rounded_rectangle([(0, 0), (width, height)], corner_radius, fill=255)
    
    # 将原图片和圆角mask结合
    output = Image.new('RGBA', (width, height), (0, 0, 0, 0))
    output.paste(img, (0, 0), mask)
    
    # 保存结果
    output.save(output_path)

def crop_to_circle(input_path, output_path):
    # 打开原始图片
    img = Image.open(input_path).convert("RGBA")
    
    # 获取图片尺寸
    width, height = img.size
    
    # 计算正方形裁剪区域的大小和位置
    size = min(width, height)
    left = (width - size) // 2
    top = (height - size) // 2
    right = left + size
    bottom = top + size
    
    # 裁剪图片为正方形
    img = img.crop((left, top, right, bottom))
    
    # 创建一个正方形画布作为遮罩
    mask = Image.new('L', (size, size), 0)
    
    # 绘制白色圆形
    draw = ImageDraw.Draw(mask)
    draw.ellipse((0, 0, size, size), fill=255)
    
    # 创建输出图像
    output = Image.new('RGBA', (size, size), (0, 0, 0, 0))
    
    # 将裁剪后的图片与圆形遮罩结合
    output.paste(img, (0, 0), mask)
    
    # 保存结果
    output.save(output_path)

# 使用示例
input_image = "/Users/allwefantasy/Downloads/我的头像_rounded.jpg"
output_image = "/Users/allwefantasy/Downloads/圆角头像.png"
crop_to_circle(input_image, output_image)
```
