```python
from autocoder.common.anything2images import Anything2Images
from autocoder.common import AutoCoderArgs
import byzerllm

byzerllm.connect_cluster()
llm = byzerllm.ByzerLLM()
llm.setup_default_model_name("deepseek_chat")
llm.setup_template("deepseek_chat", "auto")

model = "gpt_4o_mini_chat"
vl_model = byzerllm.ByzerLLM()
vl_model.setup_default_model_name(model)
vl_model.setup_template(model, "auto")

llm.setup_sub_client("vl_model", vl_model)
args = AutoCoderArgs(output="/tmp/jack", single_file=False)
v = Anything2Images(llm=llm, args=args)

@byzerllm.prompt()
def translate(prompt)->str:
    '''
    翻译下面的HTML的内容为日语: 
    
    {{ prompt }}

    然后重新输出 html, 确保布局不会变化，并且文字内容是完整的。
    输出的html必须使用 <html></html> 包裹。
    '''

def extract_html_content(html):
    import re
    pattern = re.compile(r'<html>(.*?)</html>', re.DOTALL)
    match = pattern.search(html)

    if match:
        return f"<html>{match.group(1).strip()}</html>"
    else:
        return html

v.anything_to_pdf(
    origin_file_path="/Users/allwefantasy/projects/web-typing-auto-complete/tests/t.pdf",
    target_file_path="/Users/allwefantasy/projects/web-typing-auto-complete/tests/t2.pdf",
    prompt_function=translate,
    extractor=extract_html_content
)
```

```python
import pyautogui
import time

def open_claude_ai():
    try:
        # 使用Command+Space打开Spotlight搜索
        pyautogui.hotkey('control', 'space')
        time.sleep(1)
        
        # 输入"Safari"并回车以打开Safari浏览器
        pyautogui.write('Safari')
        pyautogui.press('enter')
        time.sleep(2)
        
        # 使用Command+L聚焦到地址栏
        pyautogui.hotkey('command', 'l')
        time.sleep(0.5)
        
        # 输入Claude.ai的网址并回车
        pyautogui.write('https://claude.ai')
        pyautogui.press('enter')
        
        print("成功打开Claude.ai网站")
        return True
    except Exception as e:
        print(f"发生错误: {str(e)}")
        return False

# 执行函数
open_claude_ai()
```
